```python
import math
import numpy as np
import datetime as dt
import statsmodels.tsa.stattools as ts
from statsmodels.tsa.stattools import coint
from statsmodels.tsa.stattools import adfuller
import statsmodels.api as sm
import pandas as pd


import Quandl as quandl

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
secs =  ('HA', 'JBLU')

secX = 'WIKI/' + secs[0]
secY = 'WIKI/' + secs[1]

stocks = pd.concat([quandl.get(secX, authtoken="WEGWnsvzofrL6gGSvskF")['Close'].ix['01/01/2010':'12/31/2016'],
                    quandl.get(secY, authtoken="WEGWnsvzofrL6gGSvskF")['Close'].ix['01/01/2010':'12/31/2016']], 
                   axis=1).dropna()

stocks.columns = [secs[0], secs[1]]

stocks = stocks.apply(np.log10, axis=1)

#stocks.head(5)
plt.figure(1, figsize = (10,4))
plt.plot(stocks.HA, label='HA')
plt.plot(stocks.JBLU, label='JBLU')
plt.legend(loc = 'upper left')

score, pvalue, _ = coint(stocks.HA,stocks.JBLU)
print 'Cointegration test p-value: ' + str(pvalue)

results_HA = ts.adfuller(stocks.HA, regression='ctt')
print 'HA ADF:', results_HA[0], results_HA[4]

results_JBLU = ts.adfuller(stocks.JBLU, regression='ctt')
print 'JBLU ADF:', results_JBLU[0], results_JBLU[4]
```
